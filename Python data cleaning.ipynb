```python
import numpy as np
import pandas as pd
```

```python
appearances=pd.read_excel(r'D:\NIIT\Capstone\ds_c8_datafiles-main\appearances.xlsx')
appearances.head()
```

```python
game_events=pd.read_excel(r'D:\NIIT\Capstone\ds_c8_datafiles-main\game_events.xlsx')
game_events.head()
```

```python
game_lineups=pd.read_excel(r'D:\NIIT\Capstone\ds_c8_datafiles-main\game_lineups.xlsx')
game_lineups.head()
```

```python
games=pd.read_excel(r'D:\NIIT\Capstone\ds_c8_datafiles-main\games.xlsx')
games.head()
```

```python
players=pd.read_excel(r'D:\NIIT\Capstone\ds_c8_datafiles-main\players.xlsx')
players.head()
```

```python
football=appearances.merge(game_events,on='game_id',how='outer')
football=football.merge(game_lineups,on='game_id',how='outer')
football=football.merge(games,on='game_id',how='outer')
football=football.merge(players,on='player_id_x',how='outer')
```

```python
football.shape
```

```python
null_counts=football.isnull().sum()
print("Null_value_counts:\n",null_counts)
```

```python
football.dtypes
```

```python
for column in football.columns:
    if football[column].dtype in ['int64','float64']:
        mean_value=football[column].mean()
        football[column].fillna(mean_value,inplace=True)
    elif football[column].dtype in ['object','datetime64[ns]']:
        mode_value=football[column].mode()[0]
        football[column].fillna(mode_value,inplace=True)
football
```

```python
football.isnull().sum()
```

```python
football.drop(['home_club_formation', 'away_club_formation'],axis=1,inplace=True)
```

```python
football.isnull().sum()
```
