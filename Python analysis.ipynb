```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'D:\NIIT\Capstone\Capstone\foot_ball.csv')
df
```

```python
# Filter for players in attack position
attacking_players = df[df['position_y'] == 'Attack']

# Check if any attacking players exist
if attacking_players.empty:
  print("No players found in attack position.")
else:
# Calculate average market value for attacking players
  average_market_value = attacking_players['market_value_in_eur'].mean()

# Display result
  print("Average market value of players in attack position:", average_market_value)
```

```python
# Calculate average market value of players
average_market_value = df['market_value_in_eur'].mean()

# Display result
print("Average market value of players:", average_market_value)
```

```python
import matplotlib.pyplot as plt
df.boxplot(column='market_value_in_eur',by='position_y')
plt.xticks(rotation=90)
plt.show()
```
